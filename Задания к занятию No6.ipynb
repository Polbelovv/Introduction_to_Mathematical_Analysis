```python
import sympy as sp

# 1) Найти производные указанных функций
x = sp.symbols('x')

# 1) y = x^3 * log2(x)
y1 = x**3 * sp.log(x, 2)
dy1 = sp.diff(y1, x)

# 2) y = -10 * arctg(x) + 7 * e^x
y2 = -10 * sp.atan(x) + 7 * sp.exp(x)
dy2 = sp.diff(y2, x)

# 3) y = (1 / sqrt(x^2)) * (2 / x^3) + sqrt(7) * x
y3 = (1 / sp.sqrt(x**2)) * (2 / x**3) + sp.sqrt(7) * x
dy3 = sp.diff(y3, x)

# 4) y = cos(1 - sqrt(x)) / (1 + sqrt(x))
y4 = sp.cos(1 - sp.sqrt(x)) / (1 + sp.sqrt(x))
dy4 = sp.diff(y4, x)

# 5) y = e^(sh(5x))
y5 = sp.exp(sp.sinh(5*x))
dy5 = sp.diff(y5, x)

# 6) y = ln((x+1)(x+3)^3) / ((x^2 + 2)(x + 4)^2)
y6 = sp.ln((x + 1) * (x + 3)**3) / ((x**2 + 2) * (x + 4)**2)
dy6 = sp.diff(y6, x)

# 7) y = (sin^2(x) / ctg(x + 1)) + (cos^2(x) / tg(x + 1))
y7 = (sp.sin(x)**2 / sp.cot(x + 1)) + (sp.cos(x)**2 / sp.tan(x + 1))
dy7 = sp.diff(y7, x)

# Вывод результатов 1)
print("1) Производные функций:")
print("dy1 =", dy1)
print("dy2 =", dy2)
print("dy3 =", dy3)
print("dy4 =", dy4)
print("dy5 =", dy5)
print("dy6 =", dy6)
print("dy7 =", dy7)

# 2) Найти производную данной функции в точке
# 1) y = ln(x) / x, x0 = e
x0_1 = sp.E
y8 = sp.ln(x) / x
dy8 = sp.diff(y8, x).subs(x, x0_1)

# 2) y = sqrt(x) / sqrt(x + 1), x0 = 9
x0_2 = 9
y9 = sp.sqrt(x) / sp.sqrt(x + 1)
dy9 = sp.diff(y9, x).subs(x, x0_2)

# Вывод результатов 2)
print("\n2) Производные функции в указанных точках:")
print("dy8 в x0 =", x0_1, ":", dy8)
print("dy9 в x0 =", x0_2, ":", dy9)

# 3) Используя логарифмическую производную, найти производные функций
# 1) y = x * ln(x)
y10 = x * sp.ln(x)
dy10 = sp.diff(y10, x)

# 2) y = (x^3 - 2)^(1/3) * (x - 1)
y11 = (x**3 - 2)**(1/3) * (x - 1)
dy11 = sp.diff(y11, x)

# 3) y = (x + 5)^4 * cos(x)
y12 = (x + 5)**4 * sp.cos(x)
dy12 = sp.diff(y12, x)

# 4) y = tg(x) * cos(x)
y13 = sp.tan(x) * sp.cos(x)
dy13 = sp.diff(y13, x)

# Вывод результатов 3)
print("\n3) Производные функций:")
print("dy10 =", dy10)
print("dy11 =", dy11)
print("dy12 =", dy12)
print("dy13 =", dy13)
```
