```python
import sympy as sp

# Определяем переменные
x = sp.symbols('x')
y = sp.Function('y')(x)

# Определяем дифференциальное уравнение
differential_eq = sp.Eq(sp.Derivative(y, x) - y, 2*x - 3)

# Решаем дифференциальное уравнение
solution = sp.dsolve(differential_eq, y)

# Выводим решение
print(solution)
```

```python
import sympy as sp

# Определяем переменные
x, y = sp.symbols('x y')

# Определяем уравнение
equation = sp.Eq(x**2 * y + x * y + 1, 0)

# Решаем уравнение относительно y
solution = sp.solve(equation, y)

# Выводим решение
print(solution)
```

```python
import sympy as sp

# Определяем переменную n
n = sp.symbols('n')

# Определяем общий член ряда
term = (n + 2) / (n**2 + n + 1)

# Исследуем сходимость ряда
convergence = sp.series(term, n, 1, n=sp.oo)

# Проверяем сходимость ряда
# Используем тест на деление
limit = sp.limit(term, n, sp.oo)

# Печатаем результаты
print(f"Член ряда: {term}")
print(f"Предел члена ряда при n -> ∞: {limit}")
print(f"Ряд {term} {'сходится' if limit == 0 else 'расходится'} при n -> ∞")
```

```python
import math

def calculate_ratio(n):
    return (n + 1) ** (n + 1) / math.factorial(n + 1) * math.factorial(n) / n ** n

def check_convergence(n_max):
    for n in range(1, n_max):
        ratio = calculate_ratio(n)
        if ratio > 1:
            print(f"Ряд расходится, так как отношение соседних членов больше 1 при n = {n}")
            return False
    print("Ряд сходится")
    return True

n_max = 1000
check_convergence(n_max)
```

```python
import math

def general_term(n):
    return n * (1 - 1/n)**(n**2)

def cauchy_test(n):
    return general_term(n)**(1/n)

# Вычисление предела
n_values = range(1, 100000)  # Большое количество значений для более точного приближения
limits = [cauchy_test(n) for n in n_values]

# Оценка предела
limit = limits[-1]  # Последнее значение в списке

print(f"Предел по тесту Коши: {limit}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Определяем функцию для n-ной частичной суммы ряда
def partial_sum(x, n_max):
    s = 0
    for n in range(1, n_max + 1):
        term = ((x - 2)**(n + 1)) / (3**n * (n + 2))
        s += term
    return s

# Задаем область для x
x_values = np.linspace(-1, 5, 400)
n_max = 100  # Максимальное число членов для суммирования

# Вычисляем частичные суммы для каждого x
y_values = [partial_sum(x, n_max) for x in x_values]

# Визуализация
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='Частичные суммы ряда')
plt.title('Визуализация сходимости ряда $\sum_{n=1}^{\infty} \frac{(x-2)^{n+1}}{3^{n}(n+2)}$')
plt.xlabel('x')
plt.ylabel('Значение частичной суммы')
plt.legend()
plt.grid(True)

# Добавим линии для границ области сходимости
plt.axvline(x=-1, color='r', linestyle='--', label='Граница сходимости')
plt.axvline(x=5, color='r', linestyle='--')

plt.show()
```
